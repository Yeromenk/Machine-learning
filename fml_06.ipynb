# Clustering

```python
df_only_numeric = df.select_dtypes(np.number)
scaler = sklearn.preprocessing.MinMaxScaler()
X = scaler.fit_transform(df_only_numeric)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Apply KMeans clustering

```python
kmeans = sklearn.cluster.KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

sns.pairplot(df, hue='cluster', diag_kind='kde')
plt.show()
```

### Analyze and visualize the results

```python
for cluster in df['cluster'].unique():
    print(f"Cluster {cluster} statistics:")
    print(df[df['cluster'] == cluster].describe())
    print("\n")
```
