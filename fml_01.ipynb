Data source: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data

Important attributes description:
* SalePrice: The property's sale price in dollars. This is the target variable that you're trying to predict.
* MSSubClass: The building class
* BldgType: Type of dwelling
* HouseStyle: Style of dwelling
* OverallQual: Overall material and finish quality
* OverallCond: Overall condition rating
* YearBuilt: Original construction date
* Heating: Type of heating
* CentralAir: Central air conditioning
* GrLivArea: Above grade (ground) living area square feet

### Select only houses built in year 2000 or later

### Select houses with 2 stories or air conditioning
* HouseStyle, CentralAir

```python
air_house = df.loc[(df['HouseStyle'] == '2Story') | (df['CentralAir'] == 'Y')]
air_house
```

# Visualization

## Scatter plot
- Visualize relationship between SalePrice and GrLivArea.
- Use scatter plot from **Matplotlib**.

```python
plt.scatter(df['GrLivArea'], df['SalePrice'])
```

### Modify figure size and add title

```python
plt.figure(figsize=(10, 6))
plt.title('Relationship between SalePrice and GrLivArea')
plt.scatter(df['GrLivArea'], df['SalePrice'])
```

### Add axis labels

```python
plt.figure(figsize=(10, 6))
plt.title('Relationship between SalePrice and GrLivArea')

plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')

plt.scatter(df['GrLivArea'], df['SalePrice'])
```

### Add colors for data points based on CentralAir value.

```python
plt.figure(figsize=(10, 6), facecolor='lightgrey')
plt.title('Relationship between SalePrice and GrLivArea')

plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')

colors= df['CentralAir'].map({'Y': 'red', 'N': 'blue'})

plt.scatter(df['GrLivArea'], df['SalePrice'], c=colors)
```

## Try to use scatterplot from **Seaborn** library for scatter plot visualization.

#### Use dataframe as source and column names for axes data

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', hue='CentralAir', palette={'Y': 'blue', 'N' : 'red'})
plt.title('Relationship between SalePrice and GrLivArea')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
```

### Resize plot and add color for markers based on CentralAir column

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', hue='CentralAir', palette={'Y': 'blue', 'N': 'red'})
plt.title('Relationship between SalePrice and GrLivArea')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
```

## Line plot
- Calculate and visualize average house price in relationship to YearBuild.

```python
avg_price = df.groupby('YearBuilt')['SalePrice'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.lineplot(data=avg_price, x='YearBuilt', y='SalePrice')
plt.title('Average house price based on YearBuilt')
plt.xlabel('YearBuilt')
plt.ylabel('Average price')
plt.show()
```

## Bar plot
- Calculate and vizualize how many of houses have CentralAir
- Use Barplot for visualization

```python
have_air = df['CentralAir'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=have_air.index, y=have_air.values)
plt.title('Number of houses with CentralAir')
plt.xlabel('CentralAir')
plt.ylabel('Number of houses')
plt.show()
```

### Visualize number of building type and if they contains air conditioning using Seaborn

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='BldgType', hue='CentralAir')
plt.title('Number of building type and CentralAir')
plt.xlabel('Building type')
plt.ylabel('Number of houses')
plt.show()
```
