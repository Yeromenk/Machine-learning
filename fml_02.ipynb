## Task 1 - Outlier detection

```python
Q1_SalePrice = df.SalePrice.quantile(0.25)
Q3_SalePrice = df.SalePrice.quantile(0.75)
IQR_price = Q3_SalePrice - Q1_SalePrice
lower_bound_price = Q1_SalePrice - 1.5*IQR_price
upper_bound_price = Q3_SalePrice + 1.5*IQR_price

Q1_GrLivArea = df.GrLivArea.quantile(0.25)
Q3_GrLivArea = df.GrLivArea.quantile(0.75)
IQR_area = Q3_GrLivArea - Q1_GrLivArea
lower_bound_area = Q1_GrLivArea - 1.5*IQR_area
upper_bound_area = Q3_GrLivArea + 1.5*IQR_area

df['Outlier'] = ((df.SalePrice < lower_bound_price) | (df.SalePrice > upper_bound_price) | (df.GrLivArea < lower_bound_area) | (df.GrLivArea > upper_bound_area))

plt.figure(figsize=(16,9))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', hue='Outlier')
plt.show()
```
